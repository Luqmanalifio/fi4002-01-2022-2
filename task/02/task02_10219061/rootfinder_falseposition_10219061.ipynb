```python
import math

def f(x):
    y=3*x*x*x+8*x*x-23*x-28
    return y

x = [i * 0.03  for i in range(0, 100 + 1)]
y = [f(i) for i in x]
```

```python
xroot=[]
yroot=[]

x0 = 2
y0 = f(x0)
xroot.append(x0)
yroot.append(y0)

x1 = 3
y1 = f(x1)
xroot.append(x1)
yroot.append(y1)

i = 2
eps = 1E-6
dx = 10
print("iter root         f(root)")
while dx > eps and i < 20:
    x0 = xroot[i-2]
    x1 = xroot[i-1]
    if i == 2:
        print(f"{i-2:04d}", f"{x0:.6E}", f"{f(x0):+.6E}")
        print(f"{i-1:04d}", f"{x1:.6E}", f"{f(x1):+.6E}")
    
    x2 =  ( x0 * f(x1) - x1 * f(x0) ) / ( f(x1) - f(x0) )
    
    xroot.append(x2)
    dx = abs(f(x2))
    
    print(f"{i:04d}", f"{x2:.6E}", f"{f(x2):+.6E}")
    i += 1
```

```python
print("root =", xroot[-1])
print("f(root) =", f(xroot[-1]))
```

```python
yroot = [f(i) for i in xroot]
```

```python
import matplotlib.pyplot as plt
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, '-', xroot, yroot, 'r.', lw=2)
plt.show()
```

```python
plt.grid()
plt.xlabel("iter")
plt.ylabel("f(x)")
plt.plot(yroot, 'r.-')
plt.show()
```
