Numerical Integral & Ordinary Differential Equation

```python
# numerical integral
import math as m

def f(x) :
    y=m.sin(x)
    return y

def rectl(f,beg,end,n) :
    integ=0
    dx=(end-beg)/n
    for i in range (n) :
        xi=beg+i*dx
        integ=integ+f(xi)*dx
    return integ
def rectm(f,beg,end,n) :
    integ=0
    dx=(end-beg)/n
    for i in range (1,n+1) :
        xi=beg+0.5*i*dx
        integ=integ+f(xi)*dx
    return integ
def rectr(f,beg,end,n) :
    integ=0
    dx=(end-beg)/n
    for i in range (1,n+1) :
        xi=beg+i*dx
        integ=integ+f(xi)*dx
    return integ
def trapez(f,beg,end,n) :
    integ=0
    dx=(end-beg)/n
    for i in range (n) :
        xi=beg+i*dx
        xf=beg+(i+1)*dx
        integ=integ+0.5*(f(xi)+f(xf))*dx
    return integ
def simpson(f,beg,end,n) :
    integ=0
    dx=(end-beg)/n
    for i in range (n) :
        xi=beg+i*dx
        xf=beg+(i+1)*dx
        xmid=beg+0.5*i*dx
        integ=integ+(1/6)*(f(xi)+4*f(xmid)+f(xf))*dx
    return integ

beg=0
end=m.pi
n=10

intrl=rectl(f,beg,end,n)
intrm=rectm(f,beg,end,n)
intrr=rectr(f,beg,end,n)
inttz=trapez(f,beg,end,n)
intsm=simpson(f,beg,end,n)
print("rectangle left point integral :",intrl)
print("rectangle mid point integral :",intrm)
print("rectangle right point integral :",intrr)
print("trapezium integral :",inttz)
print("simpson integral :",intsm)
```

```python
# finite difference method
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

def fdm(n) :
    h = (5-0) / n

    # Get A
    A = np.zeros((n+1, n+1))
    A[0, 0] = 1
    A[n, n] = 1
    for i in range(1, n):
        A[i, i-1] = 1
        A[i, i] = -2
        A[i, i+1] = 1

    #print(A)

    # Get b
    b = np.zeros(n+1)
    b[0]=10
    b[1:-1] = -10*h**2
    b[-1] = 20
    #print(b)

    # solve the linear equations
    y = np.linalg.solve(A, b)

    t = np.linspace(0, 5, (n+1))

    plt.figure(figsize=(10,8))
    plt.plot(t, y)
    #plt.plot(5, 50, 'ro')
    plt.xlabel('time (s)')
    plt.ylabel('altitude (m)')
    plt.show()

print('grafik n=10') 
fdm(10)
print('grafik n=20') 
fdm(20)
print('grafik n=40') 
fdm(40)
print('grafik n=80') 
fdm(80)
print('grafik n=160') 
fdm(160)
```
